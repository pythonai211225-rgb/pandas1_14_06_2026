```python
# Inventory for Tel Aviv Branch
tel_aviv = pd.Series({
    'MacBook Pro': 5, 
    'LG Screen 27': 1, 
    'Logitech Mouse': 12
})

# Inventory for Haifa Branch
haifa = pd.Series({
    'MacBook Pro': 3, 
    'iPad Air': 7, 
    'Logitech Mouse': 8
})


# ADD another store

# 1 add another store with random items from store 1 , and store 2 and random stock
# 2 input product name from user 
# 3 print the TOTAL items in stock from all 3 stores for the product the user input --> if not exist return 0
# hint: use add + fill_value and is in 
# example:
# first  = pd.Series({'Japan': 100, 'China': 450, 'USA': 250})
# second = pd.Series({'Brazil': 70, 'China': 350, 'USA': 200})
# result = first.add(second, fill_value=0)

modiin = pd.Series ({
    'MacBook Pro': 1, 
    'LG Screen 27': 2,
    'iPad Air': 4
})

total_stock = tel_aviv.add(haifa, fill_value=0).add(modiin, fill_value=0)
print(total_stock)
product = input('product?')
if product.strip() not in total_stock:  # if product.strip() not in total_stock.keys():
    print(0)
else:
    print(total_stock[product.strip()])
    
```
