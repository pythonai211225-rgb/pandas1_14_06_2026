```python
years = [1776, 1867, 1821]
label_index = ['USA', 'CANADA', 'MEXICO']
countries = pd.Series(data=years, index=label_index) 

print(countries)
print(countries.keys())
print(countries.keys()[0])
print(list(countries.keys()))

# countries[KEY] == countries.loc[KEY]

# countries[0]  # bad -- because the index is Label and not number
countries.iloc[0]  # iloc = index location 
countries.loc['CANADA']
# countries.loc[1]  # error - no index 1 -- the index == Index(['USA', 'CANADA', 'MEXICO'], dtype='object')
```

```python
first  = pd.Series({'Japan': 100, 'China': 450, 'USA': 250})
second = pd.Series({'Brazil': 70, 'China': 350, 'USA': 200})

result = first + second
print(result)
```

```python
first  = pd.Series({'Japan': 100, 'China': 450, 'USA': 250})
second = pd.Series({'Brazil': 70, 'China': 350, 'USA': 200})

result = first.add(second, fill_value=0)
print(result)
```

```python
first  = pd.Series({'Japan': 100.2, 'China': 450, 'USA': 250})
print(first.dtype)  # float64

first  = pd.Series({'Japan': 100, 'China': 450, 'USA': 250})
print(first.dtype)  # int64

print(first)
```
